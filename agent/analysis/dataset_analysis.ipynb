```python
import glob
import json
import os
import statistics

# Find all extracted_compilation_errors.json in dataset/*/ directories
pattern = os.path.join("/Users/anon/Projects/thesis-submission/agent/dataset", "**", "extracted_compilation_errors.json")
_error_files = glob.glob(pattern)


def do_stats(error_files):
  affected_file_counts = []

  print(f"Found {len(error_files)} error files.")

  for file_path in error_files:
    # print(f"Processing {file_path}")
    try:
      with open(file_path, "r") as f:
        data = json.load(f)
        # Count the affected files (keys in the JSON)
        if len(data) > 1:
          print(file_path, data)
        affected_file_counts.append(len(data))
    except Exception as e:
      print(f"Error processing {file_path}: {e}")

  if affected_file_counts:
    mean_val = statistics.mean(affected_file_counts)
    median_val = statistics.median(affected_file_counts)
    one_file_count = affected_file_counts.count(1)
    print(f"Number of files with exactly one affected file: {one_file_count} ({one_file_count / len(affected_file_counts) * 100:.2f}%)")
    stdev_val = statistics.stdev(affected_file_counts) if len(affected_file_counts) > 1 else 0

    print(f"Mean affected files: {mean_val}")
    print(f"Median affected files: {median_val}")
    print(f"Standard deviation: {stdev_val}")
  else:
    print("No data found.")

do_stats(_error_files)

```

```python
pattern2 = os.path.join("/Users/anon/Projects/thesis-submission/zero-shot/dataset", "**", "extracted_compilation_errors.json")
error_files2 = glob.glob(pattern2)
do_stats(error_files2)
```
